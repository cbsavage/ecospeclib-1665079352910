```python
path = 'remove data from file'
files = os.listdir(path)
for file in files:
    if file.startswith('mineral'):
        f = open(os.path.join(path, file), 'r')
        lines = f.readlines()
        try:
            ind = lines.index('\n')
        except:
            print(file)
        text = lines[:ind]
        data = lines[ind+1:]
        f.close()
        f = open(os.path.join(path, file), 'w')
        for d in data:
            f.write(d)
        f.close()
        f = open(os.path.join(path, file+'_text'), 'w')
        for t in text:
            f.write(t)
        f.close()
```

```python
for file in files:
    if file.endswith('.txt_text'):
        newName = file.replace('.txt_text', '_text.txt')
        os.rename(os.path.join(path, file), os.path.join(path, newName))
```
